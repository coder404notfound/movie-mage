```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 1. Vectorize the 'genres' column
count_vectorizer_genres = CountVectorizer(stop_words='english')
genres_matrix = count_vectorizer_genres.fit_transform(df['genres'].apply(lambda x: ' '.join(x)))

# 2. Vectorize the 'cast' column
count_vectorizer_cast = CountVectorizer(stop_words='english')
cast_matrix = count_vectorizer_cast.fit_transform(df['cast'].apply(lambda x: ' '.join(x)))

# 3. Vectorize the 'tags' column
count_vectorizer_tags = CountVectorizer(stop_words='english')
tags_matrix = count_vectorizer_tags.fit_transform(df['tags'].apply(lambda x: ' '.join(x)))
```

```python
# 1. Cosine similarity for genres
genres_cosine_sim = cosine_similarity(genres_matrix, genres_matrix)

# 2. Cosine similarity for cast
cast_cosine_sim = cosine_similarity(cast_matrix, cast_matrix)

# 3. Cosine similarity for tags
tags_cosine_sim = cosine_similarity(tags_matrix, tags_matrix)
```

```python
def recommend_movies(input_value, df, cast_cosine_sim, tags_cosine_sim, genres_cosine_sim, top_n=5):
    """
    Recommend movies based on a movie title.
    Displays recommendations for Similar Cast, Similar Tags, and Similar Genres (movie titles only).
    """
    input_value = input_value.lower().strip()  # Normalize input

    # Check if input matches a movie title
    movie_idx = df[df['title'].str.lower() == input_value].index.tolist()
    if not movie_idx:
        return f"Movie '{input_value}' not found in the dataset!"

    movie_idx = movie_idx[0]

    # Prepare recommendations for Similar Cast, Similar Tags, and Similar Genres
    recommendations = {
        'Similar Cast': cast_cosine_sim[movie_idx],
        'Similar Tags': tags_cosine_sim[movie_idx],
        'Similar Genres': genres_cosine_sim[movie_idx]
    }

    result = {}
    for basis, cosine_sim in recommendations.items():
        # Get similarity scores for the selected movie
        sim_scores = list(enumerate(cosine_sim))
        # Sort by similarity score (descending)
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        # Get top N recommendations (excluding the input movie itself)
        sim_scores = sim_scores[1:top_n + 1]
        movie_indices = [x[0] for x in sim_scores]
        # Sort by popularity
        recommended_movies = df.iloc[movie_indices].sort_values(by='popularity', ascending=False)
        # Store the recommendations (titles only)
        result[basis] = recommended_movies['title'].tolist()

    return result
```

```python
# Test the function with a movie title
input_value = input("Enter a movie title: ")
recommended_movies = recommend_movies(input_value, df, cast_cosine_sim, tags_cosine_sim, genres_cosine_sim, top_n=5)

# Display recommendations
if isinstance(recommended_movies, str):
    print(recommended_movies)  # Show error message if movie not found
else:
    for basis, movies in recommended_movies.items():
        print(f"\nRecommendations Based on {basis}:\n")
        print("\n".join(movies))
```
